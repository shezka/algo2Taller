```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
df1 = pd.read_csv('dato.csv', sep=',', engine='python')
print(df1.columns.tolist())
ax1 = df1.plot(kind='scatter', x='long', y='tiempo', color='r', grid=True, label="Acierto H. Constr.")
ax1.set_ylabel("Prom. acierto (%)")
plt.savefig('dato.png')
plt.show()
```

```python
df4 = pd.read_csv('dato.csv', sep=', ', engine='python')
# print(df5)
sns.jointplot(df4['long'], df4['tiempo'], kind="reg")
plt.xlabel(r'$n$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('dato2.png')
plt.show()
```
