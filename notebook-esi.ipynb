```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
df1 = pd.read_csv('esidatos.csv', sep=',', engine='python')
# print(df1)
ax1 = df1.plot(kind='scatter', x='long', y='tiempoConPodas', color='r', logy=True, grid=True, label="Algoritmo 1 con podas")
ax1.set_ylabel("tiempo (ns)")
plt.savefig('zzzzUltmoPeorCaso.png')
plt.show()
```

```python
df4 = pd.read_csv('esidatos.csv', sep=',', engine='python')
# print(df4)
df4['long'] = df4['long'].apply(lambda x: 2**(x/2) * x * math.log(x,2))
sns.jointplot(df4['long'], df4['tiempoConPodas'], kind="reg")
plt.xlabel(r'$2^{n/2} * n * log(n)$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('zzzzAlgoritmoPeorCaso.png')
plt.show()
```
