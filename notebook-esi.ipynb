```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
df1 = pd.read_csv('esidatos.csv', sep=',', engine='python')
# print(df1)
ax2 = df1.plot(kind='scatter', x='long', y='tiempoConPodas', color='r', logy=True, grid=True, label="Algoritmo 1 con podas")
ax1 = df1.plot(kind='scatter', x='long', y='tiempoSinPodas', color='b',logy=True,  grid=True, label="Algoritmo 1 sin podas", ax=ax2)
ax1.set_ylabel("tiempo (ns)")
plt.savefig('Comparacionconmuchasopiniones.png')
plt.show()
```

```python
df4 = pd.read_csv('esidatos.csv', sep=',', engine='python')
# print(df4)
df4['long'] = df4['long'].apply(lambda x: x**3)
sns.jointplot(df4['long'], df4['tiempo'], kind="reg")
plt.xlabel(r'$x * a * log x$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('algoritmoQUIEROTERMINAR.png')
plt.show()
```
