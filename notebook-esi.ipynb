```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
df1 = pd.read_csv('esidatos.csv', sep=',', engine='python')
print(df1.columns.tolist())
ax1 = df1.plot(kind='scatter', x='long', y='tiempo', color='r', grid=True, label="Algoritmo 1 con podas")
ax1.set_ylabel("tiempo (ns)")
plt.savefig('AlgoritmoTodosOpinanMalDeTodos.png')
plt.show()
```

```python
df4 = pd.read_csv('esidatos.csv', sep=',', engine='python')
# print(df4)
df4['long'] = df4['long'].apply(lambda x: x**3)
sns.jointplot(df4['long'], df4['tiempo'], kind="reg")
plt.xlabel(r'$a * x$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('TodosOpinanMalDeTodos.png')
plt.show()
```
